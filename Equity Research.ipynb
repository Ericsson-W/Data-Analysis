```python
import pandas as pd
import math
import pandas.io.formats.style as pds
import numpy as np

pd.set_option('float_format', '{:.3f}'.format)
```

```python
# import data GS
GS = pd.read_excel(r'EPS_CHANGE_20221028_GS.xlsx')
GS = GS.loc[3:]
GS = GS.drop(GS.columns[[0,14]], axis=1)
GS.columns =['Company Name', 'BBG Ticker', 'Forecast Year', 'EPS % change','EPS GS (new)','EPS GS (old)','GS P/E FY'
            ,'GS vs. Cons EPS','Comments','% Change Sales','% Change OP','% Change PTP','Rating','Region','Conviction Sector',
             'Rating (*for CL)','Market cap (€ mn)']
GS['Currency'] = np.nan
# Remove duplicate rows
GS=GS[GS['Forecast Year'].notna()]
GS.index = range(1, len(GS.index) + 1)

for i in range(0,len(GS.index)):
    for j in range(0,len(GS.axes[1])):
        if str(GS.iat[i,j])=='nan':
            GS.iat[i,j]=GS.iat[i-1,j]
            
# Separate BBG Ticker and Currency
for k in range(0,len(GS.axes[0])):
    if len(str(GS.iat[k,1]))==3:
        GS.iat[k,17]=GS.iat[k,1]
        GS.iat[k-1,17]=GS.iat[k,1]
        
for k in range(0,len(GS.axes[0]),2):
    if len(str(GS.iat[k,1]))!=3:
        GS.iat[k+1,1]=GS.iat[k,1]

# Convert back to %
GS['EPS % change']=GS['EPS % change']*100
GS['% Change Sales	']=GS['% Change Sales']*100
GS['% Change OP']=GS['% Change OP']*100
GS['% Change PTP']=GS['% Change PTP']*100

GS
```

```python
# import data JPM
JPM = pd.read_excel(r'EPS_CHANGE_20221028_JPMCAZ.xlsx')
JPM  = JPM.loc[4:]
JPM  = JPM.drop(JPM.columns[11], axis=1)
JPM.columns = ['Company Name','BBG Ticker','Forecast Year', 'EPS % change','EPS JPM (new)','EPS Cons (mean)'
              ,'% diff vs. consensus','EPS JPM (old)','Cons PER (x)','JPMe EBIT % Change','Comments']
JPM['Currency'] = np.nan

# Remove duplicate rows
for i in range(0,len(JPM.index)):
    for j in range(0,len(JPM.axes[1])):
        if str(JPM.iat[i,j])=='nan':
            JPM.iat[i,j]=JPM.iat[i-1,j]
        if JPM.iat[i,0] == 'Valeo SA': # Valeo and Valeo SA are the same 
            JPM.iat[i,0]='Valeo'

# Separate BBG Ticker and Currency
for k in range(0,len(JPM.axes[0])):
    if len(str(JPM.iat[k,1]))==3:
        JPM.iat[k,11]=JPM.iat[k,1]
        JPM.iat[k-1,11]=JPM.iat[k,1]
        
for k in range(0,len(JPM.axes[0]),2):
    if len(str(JPM.iat[k,1]))!=3:
        JPM.iat[k+1,1]=JPM.iat[k,1]


JPM.index = range(1, len(JPM.index) + 1) 

# Convert back to %
JPM['EPS % change']=JPM['EPS % change']*100
JPM['% diff vs. consensus']=JPM['% diff vs. consensus']*100
JPM['JPMe EBIT % Change']=JPM['JPMe EBIT % Change']*100

JPM
```

```python
# import data Intermonte
Int = pd.read_excel(r'EPS_CHANGES_20221028_INTERMONTE.xlsx',sheet_name='Sheet1')
Int = Int.iloc[1:]
Int.columns = ['Company Name','BBG Ticker','Current Price (EUR)','Recommendation and TP Old','Recommendation and TP New','Forecast Year',
               'EPS INT (old)','EPS INT (new)','EPS % change','EPS Consensus','EPS % Diff','PE INT','PE Consensus']
             
Int['Currency'] = np.nan

# Remove duplicate rows
for i in range(0,len(Int.index)):
    if str(Int.iat[i,0])=='nan':
        Int.iat[i,0]=Int.iat[i-1,0]
        
for i in range(0,len(Int.index)):
    for k in range(0,len(Int.axes[0])):
        if str(Int.iat[i,k]) == 'nm': # replace nm values
            Int.iat[i,k]=0

# Separate BBG Ticker and Currency
for k in range(0,len(Int.axes[0])):
    if len(str(Int.iat[k,1]))==3:
        Int.iat[k,13]=Int.iat[k,1]
        Int.iat[k-1,13]=Int.iat[k,1]

    
for k in range(0,len(Int.axes[0]),2):
    if len(str(Int.iat[k,1]))!=3:
        Int.iat[k+1,1]=Int.iat[k,1]
        
# Convert back to %
Int['EPS % change']=Int['EPS % change']*100
Int['EPS % Diff']=Int['EPS % Diff']*100

Int
```

```python
# import data Intermonte
MS = pd.read_excel(r'EPS_CHANGES_20221028_MS.xls',sheet_name='Main2')

MS.columns=['Company Name','Ticker','Forecast Year','EPS % Change','EPS MSe','EPS Consensus','EPS % Diff','P/E Consensus'
           ,'Details','Sales % Change','Op Profit % Change']
MS = MS.iloc[1:]
MS['Currency'] = np.nan

# Remove duplicate rows
for i in range(0,len(MS.index)):
    for j in range(0,len(MS.axes[1])):
        if str(MS.iat[i,j])=='nan':
            MS.iat[i,j]=MS.iat[i-1,j]
            
# Separate BBG Ticker and Currency
for k in range(0,len(MS.axes[0])):
    if len(str(MS.iat[k,1]))==3:
        MS.iat[k,11]=MS.iat[k,1]
        MS.iat[k-1,11]=MS.iat[k,1]
        
for k in range(0,len(MS.axes[0]),2):
    if len(str(MS.iat[k,1]))!=3:
        MS.iat[k+1,1]=MS.iat[k,1]
        
# Convert back to %
MS['EPS % Change']=MS['EPS % Change']*100
MS['EPS % Diff']=MS['EPS % Diff']*100
MS['Sales % Change']=MS['Sales % Change']*100
MS['Op Profit % Change']=MS['Op Profit % Change']*100


MS
```

```python
company_list = np.append(GS['Company Name'].values,JPM['Company Name'].values)
company_list = np.append(company_list,Int['Company Name'].values)
company_list = np.append(company_list,MS['Company Name'].values)

unique_list = np.unique(company_list)

unique_list
```

```python
# Combine datasets
d = ['Broker','Company Name','Ticker','Currency','y1_EPS Chg %','y1_new EPS','y1 new EPS difference to consensus',
    'y2_EPS Chg %','y2_new EPS','y2 new EPS difference to consensus',
    'y3_EPS Chg %','y3_new EPS','y3 new EPS difference to consensus',
    'y1_PE','y2_PE','y3_PE',
    'y1_sales chg %','y1_EBIT chg %',
    'y2_sales chg %','y2_EBIT chg %',
    'y3_sales chg %','y3_EBIT chg %',
    'comments','Market Cap mn']

# y1: 2023, y2: 2024, y3:2025
df = pd.DataFrame(columns=d)
df['Company Name']=unique_list

# fill in by matching company names with other data sets
# GS
for i in range(0,len(unique_list)):
    for j in range(0,len(GS.index)):
        if df.iat[i,1] == GS.iat[j,0]: # found in GS data
            df.iat[i,0] = 'GS' # Broker
            df.iat[i,2] = GS.iat[j,1] # Ticker
            df.iat[i,3] = GS.iat[j,17] # Currency
        
            
            if GS.iat[j,2]== '2023E': # 2023 corresponds to year 1
                df.iat[i,4] = GS.iat[j,3] #y1_EPS Chg
                df.iat[i,5] = GS.iat[j,4] #new EPS
                df.iat[i,6] = GS.iat[j,7] #new EPS difference to consensus
                df.iat[i,13] = GS.iat[j,6] #y1_PE
                df.iat[i,16] = GS.iat[j,9] #y1_Sales Chg
                df.iat[i,22] = GS.iat[j,8] # Comments
                df.iat[i,23] = GS.iat[j,16] # Market Cap

# JPM
    for j in range(0,len(JPM.index)):
        if df.iat[i,1] == JPM.iat[j,0]: # found in JPM data
            if str(df.iat[i,0])=='nan' and str(JPM.iat[j,2])=='2023': # Not found in other datasets
                df.iat[i,0] = 'JPM' # Broker
                df.iat[i,2] = JPM.iat[j,1] # Ticker
                df.iat[i,3] = JPM.iat[j,11] # Currency 
                df.iat[i,4] = JPM.iat[j,3] #y1_EPS Chg
                df.iat[i,5] = JPM.iat[j,4] #new EPS
                df.iat[i,6] = JPM.iat[j,6] #new EPS difference to consensus
                df.iat[i,17] = JPM.iat[j,9] #y1_EBIT chg
                df.iat[i,22] = JPM.iat[j,10] # Comments            
            elif str(df.iat[i,0])!='nan' and str(JPM.iat[j,2])=='2023': # found in JPM data and in another set
                df.loc[-1,:]=['nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan']
                df.iat[-1,0] = 'JPM' # Broker
                df.iat[-1,1] = df.iat[i,1]
                df.iat[-1,2] = JPM.iat[j,1] # Ticker
                df.iat[-1,3] = JPM.iat[j,11] # Currency 
                df.iat[-1,4] = JPM.iat[j,3] #y1_EPS Chg
                df.iat[-1,5] = JPM.iat[j,4] #new EPS
                df.iat[-1,6] = JPM.iat[j,6] #new EPS difference to consensus
                df.iat[-1,17] = JPM.iat[j,9] #y1_EBIT chg
                df.iat[-1,22] = JPM.iat[j,10] # Comments 

                
# Intermonte
    for j in range(0,len(Int.index)):
        if df.iat[i,1] == Int.iat[j,0]: # found in Intermonte data
            if str(df.iat[i,0])=='nan' and str(Int.iat[j,5])=='2023': # Not found in other datasets
                df.iat[i,0] = 'Intermonte' # Broker
                df.iat[i,2] = Int.iat[j,1] # Ticker
                df.iat[i,3] = Int.iat[j,13] # Currency 
                df.iat[i,4] = Int.iat[j,8] #y1_EPS Chg
                df.iat[i,5] = Int.iat[j,7] #new EPS
                df.iat[i,6] = Int.iat[j,10] #new EPS difference to consensus
                df.iat[i,13] = Int.iat[j,11] # y1_PE
            elif str(df.iat[i,0])!='nan' and str(Int.iat[j,5])=='2023': # found in Intermonte data and in another set
                df.loc[-1,:]=['nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan','nan']
                df.iat[-1,0] = 'Intermonte' # Broker
                df.iat[-1,2] = Int.iat[j,1] # Ticker
                df.iat[-1,3] = Int.iat[j,13] # Currency 
                df.iat[-1,4] = Int.iat[j,8] #y1_EPS Chg
                df.iat[-1,5] = Int.iat[j,7] #new EPS
                df.iat[-1,6] = Int.iat[j,10] #new EPS difference to consensus
                df.iat[-1,13] = Int.iat[j,11] # y1_PE
                
# MS
    for j in range(0,len(MS.index)):
        if df.iat[i,1] == MS.iat[j,0]: # found in MS data
            if str(df.iat[i,0])=='nan' and str(MS.iat[j,2])=='2023': # Not found in other datasets
                df.iat[i,0] = 'MS' # Broker
                df.iat[i,2] = MS.iat[j,1] # Ticker
                df.iat[i,3] = MS.iat[j,11] # Currency
                df.iat[i,4] = MS.iat[j,3] #y1_EPS Chg
                df.iat[i,5] = MS.iat[j,4] #new EPS
                df.iat[i,6] = MS.iat[j,5] #new EPS difference to consensus
            elif str(df.iat[i,0])!='nan' and str(MS.iat[j,2])=='2023': # found in Intermonte data and in another set
                df.iat[i,0] = 'MS' # Broker
                df.iat[i,2] = MS.iat[j,1] # Ticker
                df.iat[i,3] = MS.iat[j,11] # Currency
                df.iat[i,4] = MS.iat[j,3] #y1_EPS Chg
                df.iat[i,5] = MS.iat[j,4] #new EPS
                df.iat[i,6] = MS.iat[j,5] #new EPS difference to consensus
df = df.sort_values('Ticker')
df.index = range(1, len(df.index) + 1)
df = df.replace('nan',np.NaN)
df = df.replace('----------------------------------------------------------------------------------------------------',np.NaN)
df
```
